## Moving Averages

```python
#importing all the essential libraries
import numpy as np
import pandas as pd
import datetime
import pandas_datareader as web
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#pulling in data from yahoo finance

start = datetime.datetime(2015,1,1)
end = datetime.datetime(2020, 1, 16)

df = web.DataReader("GOOGL", "yahoo", start, end)
df.head()
```

## Calculating the Moving Average

1. Moving Averages smoothens out the price action; filters noice.

2. Calculated based on past data

3. It has got two types
    a. Simple Moving Average: simple average over specific window
    b. Exponential Moving Average
    
4. Purpose: Shows the direction of Trends in data

5. Formula: (A1 + A2 + A3 +......+ An)/n
            A: Average in period
            n: Number of time period

```python
#creating a dataframe with only the needed features
#only 'Close' column is needed right now

googl_close = pd.DataFrame(df.Close)
googl_close.head()
```

## rolling method 
For calculating Moving Averages

```python
#calculating a short term Moving Average with a moving window of 20 days in the data
googl_close['MA_20Days'] = googl_close.Close.rolling(window = 20).mean()

#calculating a long term Moving Average with a moving window of200 days in the data
googl_close['MA_200Days'] = googl_close.Close.rolling(window = 200).mean()
```

## Plotting the short and long term Moving Averages

```python
plt.figure(figsize = (15,10))
plt.grid(True)
plt.plot(googl_close['Close'], label = 'close')
plt.plot(googl_close['MA_20Days'], label = '20 Days MA')
plt.plot(googl_close['MA_200Days'], label = '200 Days MA')
plt.legend()
```

## This Plot shows that the long term moving averages are much more smoother.

## The short term plots are much more sensitive than the long term

For the stated case, the 200 days moving average more generalizingly depicts the UPTREND
